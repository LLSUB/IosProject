```python
import googlemaps

gmaps = googlemaps.Client(key='')
```

```python
import pandas as pd
import time

## csv 파일 열기
filename = "대구광역시_서구_쓰레기통 설치 현황_20201125.csv"
# 대구광역시_달성군_폐형광등 및 폐건전지 분리수거함 현황_20191127.csv
# 대구광역시_서구_쓰레기통 설치 현황_20201125.csv
# 대구광역시_서구_폐건전지 및 폐형광등 분리수거함 위치_20201016.csv
# 대구광역시_중구_폐형광등 및 폐건전지 분리수거함 현황_20191129.csv
# 대구광역시_달성군_쓰레기통현황_20191127.csv
df = pd.read_csv('./csv/' + filename, encoding='cp949')

df
```

```python
addresses = df['설치위치']
addresses
```

```python
address = []
lattitude = []
longitude = []

for add in addresses:

    ## 위도, 경도 변환
    geocode_result = gmaps.geocode((add), language='ko')
    
    if geocode_result :
        lat = geocode_result[0]['geometry']['viewport']['northeast']['lat']
        lng = geocode_result[0]['geometry']['viewport']['northeast']['lng']
    else:
        lat = 0
        lng = 0
    
    print(lat, lng)
    
    address.append(add)
    lattitude.append(lat)
    longitude.append(lng)
    
## 데이터 저장
output_df = pd.DataFrame({'address': address, 'lattitude': lattitude, 'longitude': longitude})
output_df.to_csv('./csv/output_' + filename, index=False, encoding='utf-8')
```
